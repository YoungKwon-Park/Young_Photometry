## Synchronization, structured raw logging, and downstream-compatible workflow design

## Overview

The first requirement of this project was not simply to collect fluorescence signals, but to establish a measurement architecture that could preserve temporal structure, support later reconstruction, and remain compatible with downstream analysis. For that reason, acquisition was treated as the starting layer of an integrated workflow rather than as an isolated recording step.

The resulting architecture connected real-time signal collection, behavioral event synchronization, and structured raw logging within a shared temporal framework. This design made it possible to preserve traceability at the point of measurement and later convert fragmented outputs into standardized datasets for MATLAB- and Python-based processing.

## Why Acquisition Had to Be Designed as a Workflow Layer

A central difficulty in multi-device measurement environments is that signal streams and behavioral events are often generated by heterogeneous hardware modules, stored in different formats, and recorded on partially disconnected timelines. When this structure is not controlled during acquisition, temporal alignment and traceability must be reconstructed afterward, which increases processing burden and reduces reliability of later comparison.

In this project, fluorescence measurements had to be interpreted together with cue presentation, reward delivery, and licking events. The key requirement was therefore not only to record signals, but to design an acquisition layer capable of preserving synchronization across heterogeneous sources while remaining robust to interrupted recording and fragmented raw outputs.

Accordingly, the acquisition workflow was defined around three operational requirements:  
1. continuous synchronization between signal and event streams,  
2. traceable raw logging even under trial-wise file segmentation, and  
3. direct compatibility with later reconstruction and standardization.

## System Architecture and Integration Logic

The acquisition strategy was designed to preserve the highest possible level of usable information at the point of measurement. Instead of reducing measurements into simplified channel outputs during acquisition, the system retained full spectral information so that signal validity, cross-range interference, and later correction requirements could be evaluated directly from the raw measurement itself.

A spectrometer-based configuration was selected as the measurement core, while a LabVIEW-based integration layer was used to unify the optical signals and behavioral events within a shared temporal structure. This integration layer performed three core functions: continuous spectral acquisition, event synchronization, and structured raw logging.

Through this architecture, heterogeneous measurement modules were not treated as separate devices to be controlled independently, but as coordinated inputs to a single acquisition workflow.

## Acquisition Outputs as the Basis for Reconstruction and Standardization

The outputs generated at the acquisition stage were not intended to remain as isolated recording files. They were deliberately structured so that downstream preprocessing could begin without loss of temporal traceability.

To support this transition, the acquisition layer produced raw outputs that could later be reconstructed into session-level datasets, aligned to behavioral events, and standardized for repeated comparison across trials and sessions. This design was essential because later stages of the project required not only signal visualization, but also dataset reconstruction, range-based quantification, contamination assessment, algorithmic correction, and extension to additional sensor pairs.

In that sense, the acquisition layer did not merely precede later analysis. It defined the data conditions under which later analysis would be possible.
